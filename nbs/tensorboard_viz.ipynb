```python
features_train = np.load('../features_train.npy')
features_test = np.load('../features_test.npy')

tsne_subset = features_train[0:1000]
```

```python
images = datasets.load_from_disk("../datasets/wiki_train")
images = images.select(list(range(1000)))['image']
```

```python
with open('logs/embeddings/feature_vecs.tsv', 'w') as fw:
    csv_writer = csv.writer(fw, delimiter="\t")
    csv_writer.writerows(tsne_subset)
```

```python
one_square_size = int(np.ceil(np.sqrt(len(tsne_subset))))
master_width = 100 * one_square_size
master_height = 100 * one_square_size

spriteimage = Image.new(
    mode='RGB',
    size=(master_width, master_height),
    color=(0,0,0) # fully transparent
)

for count, image in enumerate(images):
    div, mod = divmod(count, one_square_size)
    h_loc = 100 * div
    w_loc = 100 * mod
    spriteimage.paste(image, (w_loc, h_loc))

spriteimage.convert('RGB').save('logs/embeddings/sprite.jpg', transparency=0)
```

```python
# other way to compute sprite image
import math 
grid = int(math.sqrt(len(images))) + 1
image_height = int(8192 / grid)         # tensorboard supports sprite images up to 8192 x 8192
image_width = int(8192 / grid)

big_image = Image.new( mode='RGB',
                       size=(image_width * grid, image_height * grid),
                       color=(0,0,0) )  # reduced it to RGB

for i in range(len(images)):
    row     = i // grid  # added integer divide
    col     = i % grid
    img     = images[i]
    img     = img.resize((image_height, image_width), Image.LANCZOS)
    row_loc = row * image_height
    col_loc = col * image_width

    big_image.paste(img, (col_loc, row_loc)) #: the order is reverse due to PIL saving
    print(row_loc, col_loc)

big_image.save('sprite.jpg') # removed transparency
```
